```python
df = df.drop(df.loc[:,'Klippemængde/kasse':'Unnamed: 11'].columns,axis=1)
```

```python
df = df.drop(df.loc[:,'Klippemængde/kasse.1':'TRUE LOG 2.1'].columns,axis=1)
```

```python
df['Liter'] = df['Liter'].str.replace(',','.').astype(float)
```

```python
df['Liter.1'] = df['Liter.1'].str.replace(',','.').astype(float)
```

```python
df
```

```python
df = df.drop(['Retning'],axis=1)
df = df.drop(['Retning.1'],axis=1)
```

```python
df
```

```python
df = df.dropna()
```

```python
df['Dato'] = pd.to_datetime(df['Dato'], format='%d/%m/%Y')

df['Week'] = df['Dato'].dt.to_period('W')

week_df = df.sort_values(by='Week')

week_df
```

```python
week_df.groupby('Week')[['Liter','Liter.1']].agg(['min','max','mean','sum'])
```

```python
grouped_df = week_df.groupby(['Hul T','Week'])['Liter'].agg('sum').reset_index()

pivot_df = grouped_df.pivot(index='Hul T',columns='Week',values='Liter')

pivot_df
```

```python
grouped_df = week_df.groupby(['Hul K','Week'])['Liter'].agg('sum').reset_index()

pivot_df = grouped_df.pivot(index='Hul K',columns='Week',values='Liter')

pivot_df
```
